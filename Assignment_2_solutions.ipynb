```python
import numpy as np
import matplotlib.pyplot as plt
```

# Generating random numbers (a:0.5, b:0.2, c:0.3)

```python
# from scratch

list_ran = []
for ix in range(1000):
    x = np.random.random()
    if x <= 0.5:
        list_ran.append("a")
    if x > 0.5 and x <= 0.7:
        list_ran.append("b")
    if x > 0.7:
        list_ran.append("c")

random_list = np.array(list_ran)
print(np.unique(random_list,return_counts = True)
```

```python
# using libraries

my_random = np.random.choice(['a','b','c'], 1000,p=[0.5,0.2,0.3])
print(np.unique(my_random,return_counts = True))
```

# Plot Random numbers as scatter plot

```python
X = np.random.rand(1000,2 )
plt.scatter(X[:,0],X[:,1],c = 'r', marker ='*')
```

# Generate Normal distribution in 2D

```python
x = np.random.multivariate_normal([1.0, 0.5], [[1.0, -1.5],[-1.5, 1.5]], 1000)
```

# Convert to Standard Normal Distribution

```python
std_x = (x - x.mean())/x.std()
```

# Plot the normal and standard normal distribution.

```python
plt.scatter(x[:,0],x[:,1])
```

```python
plt.scatter(std_x[:,0], std_x[:,1], c='b')
```

# Plotting all figures together in a single cell

```python
# fig.add_subplot(x,y,z)
# here x= no of rows, y = no of columns, z = index of figure
```

```python
#  Example 1

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1 = plt.scatter(x[:,0],x[:,1])

ax2 = fig.add_subplot(1,2,2)
ax2 = plt.scatter(std_x[:,0], std_x[:,1], c='y')

plt.show()
```

```python
#  Example 2

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1 = plt.scatter(x[:,0],x[:,1])

ax2 = fig.add_subplot(2,1,2)
ax2 = plt.scatter(std_x[:,0], std_x[:,1], c='y')

plt.show()
```
