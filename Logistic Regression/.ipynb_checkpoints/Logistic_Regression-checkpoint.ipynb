```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X_data,y_data = make_classification(n_samples = 1000, n_features = 2,n_informative = 2, n_redundant = 0,n_classes = 2,n_clusters_per_class=1, random_state=7)
```

```python
plt.scatter(X_data[:,0],X_data[:,1],c=y_data)
```

```python
ones = np.ones([1000,1])

X_data = np.hstack([ones,X_data])
```

```python
def hypothesis(x,theta):
    
    ## z = theta[0] + theta[1]*x[0] + theta[2]*x[1] but its not viable if there are many features
    z = np.dot(x,theta.T)   ## using matrix multiplication
    return 1/(1+np.exp(-z))   ## sidnoid fuction
```

```python
def likelihood(X,y,theta):
    
    m = X.shape[0]
    l = 0.0
    
    for ix in range(m):
        
        hx = hypothesis(X[ix],theta)
        l += y[ix]*np.log(hx) + (1-y[ix])*np.log(1-hx)
        
    return l
```

```python
def gradients(X,y,theta):
    
    m = X.shape[0] ## no. of rows
    n = X.shape[1] ## no. of col
    grad = np.zeros([n,])
    
    for ix in range(m):
        
        for jx in range(n):
            
            hx = hypothesis(X[ix],theta)
#             print(hx.shape)
            grad[jx] += (y[ix]-hx)*X[ix][jx]
    
    return grad/n
```

```python
def Grad_Ascent(X,y,learning_rate=0.01,threshold=0.0004):
    
    m = X.shape[0]
    n = X.shape[1]
    
    theta = np.random.randn(n,)
    ll = []
    
    for epoch in range(100):
            
        grad = gradients(X,y,theta)
        
        for ix in range(n):
            
            theta[ix] = theta[ix] + learning_rate*grad[ix]
            ll.append(likelihood(X,y,theta))
        
    return theta,ll
```

```python
theta,ll = Grad_Ascent(X_data,y_data)
```
