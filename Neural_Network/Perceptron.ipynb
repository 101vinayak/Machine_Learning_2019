## Model and helper Functions

```python
def sigmoid(z):
    
    return (1.0)/(1+np.exp(-z))
```

```python
z = np.array([1,2,3,4,5])
sigmoid(z)  ## this technique of passing a whiole array is called braodcasting, taking each element at a time
```

## Implementing Perceptron learning Algo
- learn the weights
- reduce the loss
- make th predictions

```python
def predict(X,weights):    ## weights is row vector from w0 to wn and x is an input matrix
    
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions  ## a vector    
```

```python
def loss(X,y,weights):      ## binary cross entropy
     
    y_pred = predict(X,weights)
    cost = np.mean(-y*np.log(y_pred) - (1-y)*np.log(1-y_pred))       ## using vector implementation no forloop is used
        
    return cost
```

```python
def update(X,y,weights,learning_rate):             ##perform updates for one epoch
    
    y_pred = predict(X,weights)
    dw = np.dot(X.T,y_pred - y)  ## each of the features in the ith row of the X will be multiplied with ith val of y(cap) - y
    
    m = X.shape[0]
    weights = weights - learning_rate*dw/(float(m))
    
    return weights
```

```python
def train(X,y,learning_rate = 0.5, maxEpoch = 100):
    
    ## modify the input X matrix to handle the bias term w0 ie, adding one col in the X at beggining with only ones
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    ##initialize weights
    weights = np.zeros(X.shape[1]) ## n+i entries
    
    ## iterate over all epochs and make updates
    for epoch in range(maxEpoch):
        
        weights = update(X,y,weights,learning_rate)
        if epoch%10==0:
            
            l = loss(X,y,weights)
            print("Epoch %d Loss %.4f"%(epoch,l))  ## checking progress after 10 teration and printing loss upto 4 decimal places
            
    return weights
```

```python
final_weights = train(X,y,learning_rate=0.1,maxEpoch=1000)
```

```python
def get_pred(X_test,weights,labels=False):
    
    if(X_test.shape[1] != weights.shape[0]):   ## we need N+1 col not n as weights has now n+1 due to bias
        
        ones = np.ones((X_test.shape[0],1))
        X_test = np.hstack((ones,X_test))
        
    prob = predict(X_test,weights)
    
    if not labels:
        return prob
    
    else:
        labels = np.zeros(prob.shape)
        labels[prob>=0.4] = 1
        return labels
```
