```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(n_samples=1000, centers=5,random_state=4)
print (X.shape)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
k = 5

colors = ['pink','blue','green','red','orange']

clusters = {}

for ix in range(k):
    center = 10.0*(2*np.random.random( (X.shape[1],) ) -1)
    points = []
    
    cluster = {
        "center" : center,
        "points" : points,
        "color" : colors[ix]        
    }
    
    clusters[ix] = cluster
    
print (clusters)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1])
for ix in range(k):
    plt.scatter(clusters[ix]["center"][0], clusters[ix]["center"][1], color = "black", marker = "*", s=100)
plt.show()
```

```python
def distance(v1,v2):
    return np.sqrt( np.sum( (v1-v2)**2 ) )
```

```python
for ix in range(X.shape[0]):
    
    dis = []
    curr_p = X[ix]
    
    for kx in range(k):
        d = distance(curr_p, clusters[kx]["center"])
        dis.append(d)
        
    assined_cluster = np.argmin(dis)
        
    clusters[assined_cluster]["points"].append(curr_p)
        
```
